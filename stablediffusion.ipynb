```python
from diffusers import StableDiffusionPipeline
import torch

# Ensure you have access to a GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_id = "runwayml/stable-diffusion-v1-5"  # Pre-trained model identifier
image_prompt = "A robot with I am smart written on its body"

# Load the pipeline and move it to the GPU
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to(device)

# Generate the image
generated_image = pipe(image_prompt).images[0]

# Save the generated image
generated_image.save("retro_cityscape.png")
print("Image generated and saved!")

# Display the generated image
display(generated_image)
```
